The oAuth 'dance':
<ol>
<li>Get a request token from twitter</li>
<li>Redirect user to twitter.com to authorize our application</li>
<li>If using a callback, twitter will redirect the user to us. Otherwise the user must manually supply us with the verifier code.</li>
<li>Exchange the authorized request token for an access token</li>
</ol>

<h3> Streaming API</h3>
In case we want to “keep the connection open”, and gather all the upcoming tweets about a particular event, the streaming API is what we need. We need to extend the StreamListener() to customise the way we process the incoming data. A working example that gathers all the new tweets with the #python hashtag (keep an eye on the file size...):

```python
from tweepy import Stream
from tweepy.streaming import StreamListener
 
class MyListener(StreamListener):
 
    def on_data(self, data):
        try:
            with open('ces18.json', 'a') as f:
                f.write(data)
                return True
        except BaseException as e:
            print("Error on_data: %s" % str(e))
        return True
 
    def on_error(self, status):
        print(status)
        return True
 
twitter_stream = Stream(auth, MyListener())
twitter_stream.filter(track=['#CES2018'])
```
